```python
import numpy as np
import clima_anom as ca
import cartopy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

## Example to obtain monthly climatology

```python
data_dir = '../data/3B42_199901_201212.nc'
data = ca.read_netcdf(data_dir,2)
lat = data['lat']
lon = data['lon']
pre = data['prec']
tim = data['time']
```

```python
pre_dictionary = ca.clima_anom(pre)
```

```python
lon2d, lat2d = np.meshgrid(lon, lat)
```

```python
cmap = plt.cm.Spectral_r
proj = ccrs.PlateCarree()

levels = np.arange(0,34,2)

fig = plt.figure(figsize=(8,17.5),dpi=150)
gs = gridspec.GridSpec(6,3)

ax = plt.subplot(gs[0, 0], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a1 = ax.contourf(lon2d, lat2d, pre_dictionary['jan']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.right_labels = False
gl.left_labels = True
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('January',fontsize=8,loc='left')
ax.set_title('Climatology',fontsize=8,loc='right')

ax = plt.subplot(gs[0, 1], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a2 = ax.contourf(lon2d, lat2d, pre_dictionary['feb']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.right_labels = False
gl.left_labels = False
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('February',fontsize=8,loc='left')
ax.set_title('1999/01 - 2012/12',fontsize=8,loc='right')
# plt.colorbar(a2)

ax = plt.subplot(gs[0, 2], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a3 = ax.contourf(lon2d, lat2d, pre_dictionary['mar']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.right_labels = False
gl.left_labels = False
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('March',fontsize=8,loc='left')
ax.set_title('3B42-v7',fontsize=8,loc='right')
# plt.colorbar(a3)

ax = plt.subplot(gs[1, 0], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a4 = ax.contourf(lon2d, lat2d, pre_dictionary['apr']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.right_labels = False
gl.left_labels = True
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('April',fontsize=8,loc='left')
# plt.colorbar(a4)

ax = plt.subplot(gs[1, 1], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a5 = ax.contourf(lon2d, lat2d, pre_dictionary['may']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.right_labels = False
gl.left_labels = False
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('May',fontsize=8,loc='left')
# plt.colorbar(a5)

ax = plt.subplot(gs[1, 2], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a6 = ax.contourf(lon2d, lat2d, pre_dictionary['jun']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.right_labels = False
gl.left_labels = False
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('June',fontsize=8,loc='left')
# plt.colorbar(a6)

ax = plt.subplot(gs[2, 0], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a7 = ax.contourf(lon2d, lat2d, pre_dictionary['jul']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.right_labels = False
gl.left_labels = True
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('July',fontsize=8,loc='left')
# plt.colorbar(a7)

ax = plt.subplot(gs[2, 1], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a8 = ax.contourf(lon2d, lat2d, pre_dictionary['ago']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.right_labels = False
gl.left_labels = False
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('August',fontsize=8,loc='left')
# plt.colorbar(a8)

ax = plt.subplot(gs[2, 2], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a9 = ax.contourf(lon2d, lat2d, pre_dictionary['sep']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.right_labels = False
gl.left_labels = False
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('September',fontsize=8,loc='left')
# plt.colorbar(a9)

ax = plt.subplot(gs[3, 0], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a10 = ax.contourf(lon2d, lat2d, pre_dictionary['oct']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = True
gl.right_labels = False
gl.left_labels = True
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('October',fontsize=8,loc='left')
# plt.colorbar(a10)

ax = plt.subplot(gs[3, 1], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a11 = ax.contourf(lon2d, lat2d, pre_dictionary['nov']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = True
gl.right_labels = False
gl.left_labels = False
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('November',fontsize=8,loc='left')
# plt.colorbar(a11)

ax = plt.subplot(gs[3, 2], projection = proj)
ax.coastlines(linestyle='-',linewidth=0.5,color='black',zorder=4)
a12 = ax.contourf(lon2d, lat2d, pre_dictionary['dec']['clim'][0,:,:], cmap = cmap, levels=levels)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = True
gl.right_labels = False
gl.left_labels = False
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
ax.set_title('December',fontsize=8,loc='left')
# plt.colorbar(a12)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, hspace=0.03, wspace=0.05)

cbaxes = fig.add_axes([0.125, 0.35, 0.776, 0.02]) 
fig.colorbar(a1, cax = cbaxes,orientation="horizontal")

plt.savefig('../fig/Monthly_Climatology.png',transparent = False,bbox_inches ='tight',pad_inches = 0.1)
```
